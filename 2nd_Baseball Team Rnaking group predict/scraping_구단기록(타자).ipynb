```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import time
import pandas as pd
```

###  타자 Basic1  데이터의 url 지정후 web page로 이동

####  web 주소 불러오기

```python
driver = webdriver.Chrome("chromedriver_83.exe")
```

```python
url = "https://www.koreabaseball.com/Record/Team/Hitter/Basic1.aspx"
driver.get(url)
```

#### html 파싱

```python
soup = BeautifulSoup(driver.page_source, 'lxml')
```

#### 테이블 헤더(thead) 정보 가져오기 

```python
thead1 = soup.find("div", id="cphContents_cphContents_cphContents_udpContent").find("thead")
th_list1 = thead1.find_all("th")

col_list1 = []
col_list1.append("연도")
for th in th_list1:
    
    a_tag = th.find("a")

    if a_tag == None:
        colName = th.text

    else:
        colName = a_tag.attrs["title"]

    col_list1.append(colName)

print(col_list1)
```

#### 테이블 내용 가져오기

```python
tr_list = soup.find('div', class_='record_result').find_all('tr')

Defend_list = []
for tr in tr_list:
    td_list = tr.find_all('td')
    
    t = []
    for td in td_list:
        t.append(td.text)
        
    Defend_list.append(t)
print(Defend_list)
```

```python
## 전체 데이터를 담을 공간
defend_list_all = []

for i in range(2,21,1):
    
    year = 2000 + i
    year_btn = '//*[@id="cphContents_cphContents_cphContents_ddlSeason_ddlSeason"]/option['+str(i)+']'
   
    next_year = driver.find_element_by_xpath(year_btn)
    next_year.click()
    
    time.sleep(1)
    
    soup = BeautifulSoup(driver.page_source, 'lxml')
    
    ## 테이블 헤더 + 바디
    record_result = soup.find("div", class_="record_result")
    tr_list = record_result.find("tbody").find_all("tr")
    
    ## 1년치 데이터를 담을 공간
    defend_list = []
    for tr in tr_list:
        td_list = tr.find_all("td")
        
        t = []
        t.append(year)
        for td in td_list:
            t.append(td.text)
            
        defend_list.append(t)
        
    print(str(year) + "년도 데이터 수집 끗!")    
    
    ## 전체 데이터 공간에 1년치 데이터 추가 
    defend_list_all.extend(defend_list)

```

```python
defend_df = pd.DataFrame(defend_list_all, columns=col_list1)
defend_df
```
